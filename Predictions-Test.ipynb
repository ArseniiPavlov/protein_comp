```python
import numpy as np
import tensorflow as tf
import InceptionV3_fine_tune2
import pandas as pd
import img_proc_scikit
from PIL import Image, ImageFile
from os import path, listdir, makedirs, fsdecode, getcwd
from skimage.transform import resize
```

```python
ImageFile.LOAD_TRUNCATED_IMAGES = True


test_df = img_proc_scikit.TrainData('sample_submission.csv').make_train(class_ID='Predicted')
img_rows, img_cols = 299, 299  # Resolution of inputs
channel = 3
num_classes = 28
batch_size = 12

model = InceptionV3_fine_tune2.InceptionV3_make('InceptionV3_ft2.h5', 28)
f1_score = InceptionV3_fine_tune2.f1

test_dir = getcwd() + '\\test'
test_df['Preds'] = None
i = 0
```

```python
for line in test_df.values:
    sample = img_proc_scikit.save_rgb_image(line[0], line[2], test_dir, labeled_dir=None, save_img=False)
    sample = resize(sample, (299, 299))
    sample = np.array(sample)
    line[3] = model.predict(np.expand_dims(sample, axis=0), batch_size=1)[0]
    i += 1
    if (i % 100) == 0:
        print('{} images were predicted'.format(i))
```

```python
import pickle
test_df.to_csv('predicted-test.csv')
with open('test_df.pkl', 'wb') as f:
    pickle.dump(test_df, f)
```
